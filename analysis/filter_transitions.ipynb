```python
import pandas as pd
import numpy as np
```

```python
fn = "../data/IDX/AX_stab_pool/generated_SRM_table.csv"
data = pd.read_csv(fn)
data
```

```python
data = data[(data["Normalized Intensity"] != float("inf")) & (data["Normalized Intensity"] > 2000)]
data
```

```python
cpds = list(set(data["Name"].values))
print(len(cpds))
```

```python
toKeep = []
annotation = {}
for cpd in cpds:
    filt = data[data["Name"] == cpd]
    filt.sort_values(by="Normalized Intensity",ascending=False)
    charge = filt.at[filt.index.values[0],"Charge"]
    filt = filt[filt["Charge"] == charge]
    toKeep += list(filt.index.values)
    mapper = ["Quant","Qual 1","Qual 2","Qual 3","Qual 4"]
    for ind,lab in zip(filt.index.values,mapper[:len(filt)]):
        annotation[ind] = {"Ion type":lab}
```

```python
data = data.loc[toKeep,:]
data = pd.concat((data,pd.DataFrame.from_dict(annotation,orient="index")),axis=1,ignore_index=False)
data
```

```python
compoundInfo = {}
for index,row in data.iterrows():
    compoundInfo[row["Name"]] = {"mz":row["mz"],"Charge":row["Charge"],"rt":np.mean([row["rt_start"],row["rt_end"]])}
```

```python
mzTol = 5
rtTol = .33
badCpds = []

for cpd in compoundInfo:
    good = True
    for cpd2 in compoundInfo:
        if cpd != cpd2 and cpd2 not in badCpds:
            if compoundInfo[cpd]["Charge"] == compoundInfo[cpd2]["Charge"]:
                if abs(compoundInfo[cpd]["rt"] - compoundInfo[cpd2]["rt"]) < rtTol:
                    if 1e6 * np.abs(compoundInfo[cpd]["mz"] - compoundInfo[cpd2]["mz"])/compoundInfo[cpd]["mz"] < mzTol:
                        good = False
                        break
    if not good:
        badCpds.append(cpd)
```

```python
print(badCpds)
```

```python
toDrop = []
for cpd in badCpds:
    filt = data[data["Name"] == cpd]
    toDrop += list(filt.index.values)
data = data.drop(toDrop)
data.sort_values(by="Name")
data
```

```python
data.to_csv("../data/IDX/AX_stab_pool/generated_SRM_table_filtered.csv")
```
